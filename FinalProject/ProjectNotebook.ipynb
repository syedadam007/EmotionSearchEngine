### Terminal

Because this project was made in the terminal, I still wanted it to be possible for my code to display visual data and musical data and this was achieved by using the pyglet module and subprocess module. Subprocess module is used to play .wav audio files and Pyglet module is used to play .gif media files.

### Instructions

You must have Python 3.x.x and above for this program to work correctly.

Pyglet Installation: <br />
https://pyglet.readthedocs.io/en/stable/programming_guide/installation.html

Type this command into your terminal: <br />
pip3 install pyglet <br />

To play the sound files in my project, you will need to download FFMPEG: <br />
https://www.ffmpeg.org/download.html

Pytest Installation: <br>
pip3 install pytest --user or pip install pytest --user

This module is used so that we can check to see if the methods defined within the classes are working correctly.

### emotions.py

This file contains the class Emotions() and the functions containing the visual and sound files that are triggered whenever an emotion is inputted by the user.

Code resource for implementing moving images within Python: (Used code from here) <br> https://codeloop.org/python-pyglet-adding-gif-image/

Subprocess documentation for implementing audio within Python: <br>
https://docs.python.org/2/library/subprocess.html

### starwars.py

This file contains the class StarWars() and functions that returns text translations of characters from the Star Wars films based on the inputted emotion by the user.

### main.py

As the name suggests, this file contains all the code that acts as the visual user interface when the system is presented to the user.

### test_translators.py

This file test to see if the methods defined in StarWars() class works. Consists of 3 tests.
